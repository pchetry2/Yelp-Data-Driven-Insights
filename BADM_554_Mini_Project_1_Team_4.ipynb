**Business Questions & Queries**

1. Which cities have the highest number of reviews for restaurants? This helps to determine the location that indicates high restaurant engagement and popularity within a specific city.

```python
%%sql
SELECT b.city, SUM(b.review_count) as total_reviews
FROM business b
JOIN categories c ON b.business_id = c.business_id
WHERE c.categories LIKE '%Restaurants%'
GROUP BY b.city
ORDER BY total_reviews DESC;
```

2. What is the number of highly rated restaurants and reviews for the top 3 cities? Helps to provide insights into customer satisfaction and potential competition in those cities.

```python
%%sql
SELECT
    b.city,
    COUNT(DISTINCT b.business_id) AS number_of_restaurants,
    COUNT(r.review_id) AS number_of_reviews
FROM
    business b
JOIN
    review r ON b.business_id = r.business_id
JOIN
    categories c ON b.business_id = c.business_id
WHERE
    c.categories LIKE '%Restaurants%' AND r.stars >= 4 -- assuming 4 and above is 'highly rated'
GROUP BY
    b.city
ORDER BY
    number_of_reviews DESC
LIMIT 3;
```

3. What are the attributes of these high-performing restaurants? The results can help in the understanding of the key factors contributing to restaurant success.

```python
%%sql
SELECT
    b.business_id,
    AVG(r.stars) as average_rating,
    COUNT(r.review_id) as review_count,
    MAX(a.RestaurantsTakeOut) as RestaurantsTakeOut,
    MAX(a.RestaurantsDelivery) as RestaurantsDelivery
    -- Add any other specific attributes here with aggregate functions if needed
FROM
    business b
JOIN
    review r ON b.business_id = r.business_id
JOIN
    attributes a ON b.business_id = a.business_id
JOIN
    categories c ON b.business_id = c.business_id
WHERE
    c.categories LIKE '%Restaurants%'
GROUP BY
    b.business_id
HAVING
    average_rating >= 4 AND review_count > 50
ORDER BY
    average_rating DESC, review_count DESC;


```

4. Is there a pattern that can be seen amongst the attributes presented in high-performing restaurants? We can use this data to analyze for any discernible patterns or trends that could help inform decisions related to restaurant setup, menu planning, marketing strategies, etc.

```python
%%sql
SELECT
    AVG(CAST(a.RestaurantsPriceRange2 AS UNSIGNED)) as average_price_range,  -- Assuming RestaurantsPriceRange2 is stored as a string that can be cast to an unsigned integer
    SUM(CASE WHEN a.BusinessAcceptsCreditCards = '1' THEN 1 ELSE 0 END) as accepts_credit_cards_count,  -- Assuming 'True' is represented as '1' in the database
    SUM(CASE WHEN a.BikeParking = '1' THEN 1 ELSE 0 END) as bike_parking_count,
    SUM(CASE WHEN a.RestaurantsTakeOut = '1' THEN 1 ELSE 0 END) as takeout_count,
    SUM(CASE WHEN a.RestaurantsDelivery = '1' THEN 1 ELSE 0 END) as delivery_count,
    COUNT(*) as total_high_performing_restaurants
FROM
    attributes a
JOIN
    (SELECT b.business_id
     FROM business b
     JOIN review r ON b.business_id = r.business_id
     JOIN categories c ON b.business_id = c.ID  -- The join condition is corrected based on your tables' structure
     WHERE c.categories LIKE '%Restaurants%'  -- The column name is corrected to 'categories'
     GROUP BY b.business_id
     HAVING AVG(r.stars) >= 4 AND COUNT(r.review_id) > 50) as HighPerformingRestaurants
ON
    a.business_id = HighPerformingRestaurants.business_id;


```

5. In which cities are the high-performing restaurant attributes missing?  Helps to identify a gap in the market, presenting an opportunity for new restaurant ventures or expansion.

```python
%%sql
WITH HighPerformingRestaurants AS (
    SELECT r.business_id
    FROM review r
    GROUP BY r.business_id
    HAVING AVG(r.stars) >= 4 AND COUNT(r.review_id) > 50
),
HighPerformingAttributes AS (
    SELECT
        a.business_id,
        AVG(a.RestaurantsPriceRange2) as average_price_range,
        MAX(a.BusinessAcceptsCreditCards) as credit_cards,
        MAX(a.BikeParking) as bike_parking
        -- Ensure any additional attributes are correctly aggregated or handled.
    FROM attributes a
    JOIN HighPerformingRestaurants hpr ON a.business_id = hpr.business_id
    GROUP BY a.business_id -- If attributes are unique per business, group by business_id.
),
CitiesWithHighPerformingRestaurants AS (
    SELECT DISTINCT b.city
    FROM business b
    JOIN HighPerformingAttributes hpa ON b.business_id = hpa.business_id
),
CitiesMissingHighPerformingAttributes AS (
    SELECT DISTINCT b.city
    FROM business b
    WHERE NOT EXISTS (
        SELECT 1
        FROM CitiesWithHighPerformingRestaurants cwhpr
        WHERE b.city = cwhpr.city
    )
    -- Assuming there's a way to filter businesses by category, such as a join with a categories table.
)
SELECT city
FROM CitiesMissingHighPerformingAttributes;


```

6. What is the average price range of highly rated restaurants in the top cities?

```python
%%sql
SELECT
    b.city,
    AVG(CAST(a.RestaurantsPriceRange2 AS UNSIGNED)) as average_price_range
FROM
    business b
INNER JOIN
    review r ON b.business_id = r.business_id
INNER JOIN
    attributes a ON b.business_id = a.business_id
INNER JOIN  -- This join has been added to include the categories table
    categories c ON b.business_id = c.business_id
WHERE
    c.categories LIKE '%Restaurants%' AND r.stars >= 4 -- Corrected to use the categories table
GROUP BY
    b.city
ORDER BY
    average_price_range DESC;


```

7. Find the cities where users usually rate restaurant businesses highly, and check whether they have a restaurant in that city for our high-performing attributes.

```python
%%sql
WITH HighlyRatedCities AS (
    SELECT
        b.city,
        AVG(r.stars) as average_stars
    FROM
        review r
    INNER JOIN
        business b ON r.business_id = b.business_id
    INNER JOIN
        categories c ON b.business_id = c.business_id
    WHERE
        c.categories LIKE '%Restaurants%'
    GROUP BY
        b.city
    HAVING
        AVG(r.stars) >= 4
),
HighPerformingRestaurants AS (
    SELECT
        a.business_id
    FROM
        attributes a
    WHERE
        a.BusinessAcceptsCreditCards = 'True'
        AND a.BikeParking = 'True'
),
CitiesWithHighPerformingRestaurants AS (
    SELECT DISTINCT
        b.city
    FROM
        business b
    INNER JOIN
        HighPerformingRestaurants hpr ON b.business_id = hpr.business_id
)
SELECT
    hrc.city
FROM
    HighlyRatedCities hrc
LEFT JOIN
    CitiesWithHighPerformingRestaurants cwhpr ON hrc.city = cwhpr.city
WHERE
    cwhpr.city IS NULL;



```

8. Find the cities where users prefer long operating hours restaurants- This analysis offers insights into consumer preferences for extended hrs demand and informs cost benefit analysis such variable overhead cost.

```python
%%sql
SELECT
    b.city,
    AVG(r.stars) as average_rating,
    AVG(
        LENGTH(COALESCE(h.Monday, '')) + LENGTH(COALESCE(h.Tuesday, '')) + LENGTH(COALESCE(h.Wednesday, '')) +
        LENGTH(COALESCE(h.Thursday, '')) + LENGTH(COALESCE(h.Friday, '')) + LENGTH(COALESCE(h.Saturday, '')) +
        LENGTH(COALESCE(h.Sunday, ''))
    ) / 7 as average_daily_operating_hours
FROM
    business b
INNER JOIN
    review r ON b.business_id = r.business_id
INNER JOIN
    hours h ON b.business_id = h.business_id
INNER JOIN
    categories c ON b.business_id = c.business_id
WHERE
    c.categories LIKE '%Restaurants%' AND r.stars >= 4 -- Using 'categories' based on your table structure
GROUP BY
    b.city
HAVING
    average_daily_operating_hours > 12 -- Replace with a specific number like 10 for hours
ORDER BY
    average_rating DESC, average_daily_operating_hours DESC;


```

9. Identify the cities where the availability of service offers such as restaurant take out or delivery service have impact on the highly rated restaurants - Helps to identify market opportunity, competitor analysis, increase operational efficiency.

```python
%%sql
SELECT
    b.city,
    AVG(r.stars) AS average_rating,
    SUM(CASE WHEN a.RestaurantsTakeOut = 'True' THEN 1 ELSE 0 END) AS takeout_offered,
    SUM(CASE WHEN a.RestaurantsDelivery = 'True' THEN 1 ELSE 0 END) AS delivery_offered,
    COUNT(*) AS total_restaurants
FROM
    business b
JOIN
    review r ON b.business_id = r.business_id
JOIN
    attributes a ON b.business_id = a.business_id
JOIN  -- Added join to the categories table
    categories c ON b.business_id = c.business_id
WHERE
    c.categories LIKE '%Restaurants%'  -- Adjusted to the correct table alias
    AND r.stars >= 4 -- Threshold for highly rated
GROUP BY
    b.city
ORDER BY
    average_rating DESC;


```
