# Which cities have the highest number of reviews for restaurants?

```python
def is_restaurant(category):
    if 'restaurant' in category.lower():
        return 1
    else:
        return 0
    
business = business.dropna(subset=['categories'], axis=0)
business['restaurant_flag'] = business.categories.apply(is_restaurant)
business_rest = business[business.restaurant_flag == 1]
business_grouped = business_rest.groupby(by='city').review_count.sum().reset_index()
business_grouped = business_grouped.sort_values(by='review_count', ascending=False)
business_grouped.index = range(len(business_grouped))
print(business_grouped)
```

# What is the number of highly rated restaurants and reviews for the top 3 cities? Helps to provide insights into customer satisfaction and potential competition in those cities.

```python
business_rest_city = business_rest[business_rest.city.isin(['Philadelphia', 'New Orleans', 'Nashville'])]
high_rated = review[review.stars >= 4]
bus_rev = business_rest_city.merge(high_rated[['review_id', 'business_id']], on='business_id', how='inner')
bus_rev_grouped = bus_rev.groupby(by='city').agg({'business_id': 'nunique', 'review_id': 'nunique'}).reset_index()
bus_rev_grouped = bus_rev_grouped.rename(columns={'business_id': 'no_of_restaurants', 'review_id': 'number_of_reviews'})
print(bus_rev_grouped)
```

# What are the attributes of these high-performing restaurants? The results can help in the understanding of the key factors contributing to restaurant success.

```python
bus_attr = bus_rev.attributes.apply(pd.Series)
df_grouped_1 = bus_rev.groupby(by='business_id').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index()
```

```python
bus_attr = pd.concat([bus_rev[['business_id']], bus_attr], ignore_index=False, axis=1)
bus_attr_grouped = bus_attr.groupby(by='business_id').agg({'RestaurantsTakeOut': 'max', 'RestaurantsDelivery': 'max'}).reset_index()
bus_attr_grouped = df_grouped_1.merge(bus_attr_grouped, on='business_id', how='inner')
bus_attr_grouped = bus_attr_grouped.rename(columns={'stars': 'avg_rating', 'review_id': 'review_counts'})
print(bus_attr_grouped)
```

# Is there a pattern that can be seen amongst the attributes presented in high-performing restaurants?

```python
bus_attr.RestaurantsPriceRange2 = bus_attr.RestaurantsPriceRange2.replace('None', '0').fillna('0')
bus_attr.RestaurantsPriceRange2 = bus_attr.RestaurantsPriceRange2.astype('int')
bus_attr.RestaurantsPriceRange2 = bus_attr.RestaurantsPriceRange2.replace(0, np.nan)
```

```python
bus_attr.BusinessAcceptsCreditCards = bus_attr.BusinessAcceptsCreditCards.replace('False', 0).replace('True', 1)
bus_attr.BikeParking = bus_attr.BikeParking.replace('False', 0).replace('True', 1).replace('None', np.nan)
bus_attr.RestaurantsTakeOut = bus_attr.RestaurantsTakeOut.replace('False', 0).replace('True', 1).replace('None', np.nan)
bus_attr.RestaurantsDelivery = bus_attr.RestaurantsDelivery.replace('False', 0).replace('True', 1).replace('None', np.nan)
```

```python
bus_attr_pattern = bus_attr.agg({
    'RestaurantsPriceRange2': 'mean', 'BusinessAcceptsCreditCards': 'sum', 'BikeParking': 'sum', 'RestaurantsTakeOut': 'sum',
    'RestaurantsDelivery': 'sum'}).reset_index()
print(bus_attr_pattern)
```

# What is the effect of restaurant category (e.g., Italian, Chinese, Vegan) on its average rating and review volume in major cities?

```python
business_rest['is_italian'] = business_rest.categories.apply(lambda x: 1 if 'italian' in x.lower() else 0)
business_rest['is_chinese'] = business_rest.categories.apply(lambda x: 1 if 'chinese' in x.lower() else 0)
business_rest['is_vegan'] = business_rest.categories.apply(lambda x: 1 if 'vegan' in x.lower() else 0)
business_rest['is_mexican'] = business_rest.categories.apply(lambda x: 1 if 'mexican' in x.lower() else 0)
business_rest['is_indian'] = business_rest.categories.apply(lambda x: 1 if 'indian' in x.lower() else 0)
business_rest['is_american'] = business_rest.categories.apply(lambda x: 1 if 'american' in x.lower() else 0)
```

```python
bus_rev_all = business_rest.merge(review[['review_id', 'business_id']], on='business_id', how='inner')

print(bus_rev_all.groupby(by='is_italian').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index().rename(
    columns={'stars': 'avg_rating', 'review_id': 'review_counts'}))
print(bus_rev_all.groupby(by='is_chinese').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index().rename(
    columns={'stars': 'avg_rating', 'review_id': 'review_counts'}))
print(bus_rev_all.groupby(by='is_vegan').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index().rename(
    columns={'stars': 'avg_rating', 'review_id': 'review_counts'}))
print(bus_rev_all.groupby(by='is_mexican').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index().rename(
    columns={'stars': 'avg_rating', 'review_id': 'review_counts'}))
print(bus_rev_all.groupby(by='is_indian').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index().rename(
    columns={'stars': 'avg_rating', 'review_id': 'review_counts'}))
print(bus_rev_all.groupby(by='is_american').agg({'stars': 'mean', 'review_id': 'nunique'}).reset_index().rename(
    columns={'stars': 'avg_rating', 'review_id': 'review_counts'}))
```
